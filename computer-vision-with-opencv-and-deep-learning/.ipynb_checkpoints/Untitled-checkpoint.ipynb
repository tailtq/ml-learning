```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
image = Image.open('course/DATA/00-puppy.jpg')
```

```python
image_arr = np.asarray(image)
```

```python
image_arr.shape
```

```python
# RED CHANNEL
red_arr = image_arr[:,:,0]
plt.imshow(red_arr, cmap='gray')
```

```python
# GREEN CHANNEL
green_arr = image_arr[:,:,1]
plt.imshow(green_arr, cmap='gray')
```

```python
# BLUE CHANNEL
blue_arr = image_arr[:,:,0]
plt.imshow(blue_arr, cmap='gray')
```
