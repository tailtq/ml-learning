```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
image = Image.open('course/DATA/00-puppy.jpg')
```

```python
image_arr = np.asarray(image)
```

```python
image_arr.shape
```

```python
# RED CHANNEL
plt.imshow(image_arr[:,:,0], cmap='gray')
```

```python
# GREEN CHANNEL
plt.imshow(image_arr[:,:,1], cmap='gray')
```

```python
# BLUE CHANNEL
plt.imshow(image_arr[:,:,2], cmap='gray')
```

```python
red_arr = image_arr.copy()
red_arr[:,:,1] = 0
red_arr[:,:,2] = 0
plt.imshow(red_arr, cmap='gray')
```
