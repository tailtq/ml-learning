```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img = cv2.imread('course/DATA/sammy_face.jpg')

display_img(img)
```

```python
edge = cv2.Canny(img, threshold1=180, threshold2=255)
display_img(edge)
```

```python
median = np.median(img)

# find threshold
lower = int(max(0, 0.7 * median))
upper = int(min(255, 1.3 * median))
```

```python
edge = cv2.Canny(img, threshold1=lower, threshold2=upper)
display_img(edge)
```

```python
blurred_img = cv2.blur(img, ksize=(5, 5))

edge = cv2.Canny(blurred_img, threshold1=lower, threshold2=upper + 50)
display_img(edge)
```
