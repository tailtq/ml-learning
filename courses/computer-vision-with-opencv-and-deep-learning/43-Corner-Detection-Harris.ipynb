```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
flat_chessboard = cv2.imread('course/DATA/flat_chessboard.png')
flat_chessboard = cv2.cvtColor(flat_chessboard, cv2.COLOR_BGR2RGB)
flat_chessboard_gray = cv2.cvtColor(flat_chessboard, cv2.COLOR_RGB2GRAY)
```

```python
real_chessboard = cv2.imread('course/DATA/real_chessboard.jpg')
real_chessboard = cv2.cvtColor(real_chessboard, cv2.COLOR_BGR2RGB)
real_chessboard_gray = cv2.cvtColor(real_chessboard, cv2.COLOR_RGB2GRAY)
```

```python
display_img(flat_chessboard, 'gray')
```

```python
display_img(real_chessboard, 'gray')
```

```python
gray = np.float32(flat_chessboard_gray)
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
dst = cv2.dilate(dst, None)
```

```python
flat_chessboard[dst > 0.01 * dst.max()] = [255, 0, 0]
```

```python
display_img(flat_chessboard)
```

```python
gray = np.float32(real_chessboard_gray)
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
dst = cv2.dilate(dst, None)
```

```python
real_chessboard[dst > 0.01 * dst.max()] = [255, 0, 0]
```

```python
display_img(real_chessboard)
```
