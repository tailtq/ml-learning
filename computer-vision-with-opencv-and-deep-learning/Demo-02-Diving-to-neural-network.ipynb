```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
from keras.callbacks import LambdaCallback
```

```python
model = tf.keras.Sequential([
    keras.layers.Dense(units=5, input_shape=[1], activation='sigmoid'),
    keras.layers.Dense(units=1)
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```python
print_weights = LambdaCallback(on_epoch_end=lambda batch, logs: print(model.layers[0].get_weights(), model.layers[1].get_weights()))
```

```python
model.fit(xs, ys, epochs=500, callbacks=[print_weights])
```

```python
print(model.predict([10.0]))
```
