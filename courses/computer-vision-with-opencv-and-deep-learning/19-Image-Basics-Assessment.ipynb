#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('course/DATA/dog_backpack.jpg')

image.shape
```

```python
plt.imshow(image)
```

```python
fix_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_image)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
flip_image = cv2.flip(fix_image, 0)
plt.imshow(flip_image)
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
new_image = fix_image.copy()

cv2.rectangle(new_image, pt1=(220, 380), pt2=(630, 730), color=(255, 0, 0), thickness=10)

plt.imshow(new_image)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
new_image = fix_image.copy()

vertices = np.array([ [430, 390], [200, 700], [600, 700] ])

vertices = vertices.reshape((-1, 1, 2))

cv2.polylines(new_image, [vertices], isClosed=True, color=(0, 0, 255), thickness=10)

plt.imshow(new_image)
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
import numpy as np
import cv2

new_image = fix_image.copy()

def draw_circle(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(new_image, (x, y), 50, color=(255, 0, 0), thickness=10)
    
cv2.namedWindow('my_drawing')

cv2.setMouseCallback('my_drawing', draw_circle)

while True:
    cv2.imshow('my_drawing', new_image)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
