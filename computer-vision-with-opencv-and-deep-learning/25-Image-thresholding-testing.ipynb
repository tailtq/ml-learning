```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img1 = cv2.imread('course/DATA/00-puppy.jpg', 0)

ret, th1 = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY)

show_pic(th1)
```

```python
img2 = cv2.imread('course/DATA/sudoku.jpg', 0)

# img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

show_pic(img2)
```

```python
ret, th2 = cv2.threshold(img2, 160, 255, cv2.THRESH_BINARY)

show_pic(th2)
```
