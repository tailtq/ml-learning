```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/plane.jpg').astype(np.float32) / 255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
img.shape
```

```python
brighter_img = np.power(img, 2)

plt.imshow(brighter_img)
```

```python
contrastive_img = (img - 0.5) * 0.3 + 0.5

plt.imshow(contrastive_img)
```
