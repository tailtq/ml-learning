```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nadia = cv2.imread('course/DATA/Nadia_Murad.jpg', 0)
denis = cv2.imread('course/DATA/Denis_Mukwege.jpg', 0)
solvay = cv2.imread('course/DATA/solvay_conference.jpg', 0)
```

```python
face_cascade = cv2.CascadeClassifier('course/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x + w, y + h), (255, 255, 255), 10)
        
    return face_img
```

```python
result = detect_face(solvay)
```

```python
plt.imshow(result, cmap='gray')
```

```python
def adj_detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.11, minNeighbors=4)
    
    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x + w, y + h), (255, 255, 255), 10)
        
    return face_img
```

```python
result = adj_detect_face(solvay)
```

```python
plt.imshow(result, cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('course/DATA/haarcascades/haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    eyes_img = img.copy()
    eyes_rects = eye_cascade.detectMultiScale(eyes_img)
    
    for (x, y, w, h) in eyes_rects:
        cv2.rectangle(eyes_img, (x, y), (x + w, y + h), (255, 255, 255), 10)
        
    return eyes_img
```

```python
result = detect_eyes(nadia)
```

```python
plt.imshow(result, cmap='gray')
```
