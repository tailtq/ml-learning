```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600, 600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text='ABCDE', org=(50, 300), fontFace=font, fontScale=5, color=(255, 255, 255), thickness=25)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
kernel = np.ones((5, 5), dtype=np.uint8)
```

```python
result = cv2.erode(img, kernel, iterations=4)

display_img(result)
```

```python
# morphological opening
white_noise = np.random.randint(0, 2, size=(600, 600)) * 255
```

```python
combination = white_noise + img
```

```python
display_img(combination)
```

```python
result = cv2.morphologyEx(combination, cv2.MORPH_OPEN, kernel)

display_img(result)
```

```python
display_img(img)
```

```python
# morphological closing
black_noise = np.random.randint(0, 2, (600, 600)) * -255
```

```python
combination = black_noise + img
combination[combination == -255] = 0
```

```python
display_img(combination)
```

```python
result = cv2.morphologyEx(combination, cv2.MORPH_CLOSE, kernel)
display_img(result)
```

```python
result = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
display_img(result)
```
